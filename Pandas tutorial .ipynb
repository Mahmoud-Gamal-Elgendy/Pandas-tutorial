### *7-Statistics methods.*
**mean**
    - The mean is the sum of the elements divided by their number
    
**median**
    - How to calculate the Median : First, we arrange the elements in ascending order, then we take the value that appears       in the middle of the elements. For example, if we have 7 elements, we take the element whose position is 4 after the       arrangement.

**mode**
    - The value is duplicated in the list, and it is possible that there is more than one mode or that there is no mode at       all

```python
Labtops_data.weight_kg.mean()
```

```python
Labtops_data.weight_kg.median()
```

```python
Labtops_data.weight_kg.mode()
```

### *8-Selection and Indexing.*

```python
# If I want get only one column from the DataFrame
Labtops_data.name # or Labtops_data["name"]
```

```python
# To select the first row of data in a DataFrame 
Labtops_data.iloc[0] # [row,column]
```

```python
# to select just the second and third entries and thier value in the first 3 columns
Labtops_data.iloc[1:3,0:3]
```

```python
# to select index based on the data index value, not its position we use loc
Labtops_data.loc[1:3,["price_eur","name"]]
```

```python
# Conditional Selection based on price_eur column
Labtops_data.loc[Labtops_data.price_eur>2000]
```

```python
# select data whose value "is in" a list of values isin
Labtops_data.display_resolution.isin(["3840x2160"])
```

### *9-handle Missing Data.*

```python
# Replace Every NaN in display_resolution Column to Unkown 
Labtops_data.display_resolution.fillna("Unkown",inplace=True)
Labtops_data
```

```python
# Replace Every 3:10 in display_ratio Column to 16:10 
Labtops_data.display_ratio.replace("3:2","16:10")
```

```python
# to count missed data in display_resolution
Labtops_data.display_resolution.fillna('Unknown').value_counts()
```

### *10-Group by .*

```python
new_lab_list=Labtops_data.groupby("display_resolution")
for dis , dis_df in new_lab_list:
    print(dis)
    print(dis_df)
```

### *11- Concat dataframs.*

```python
Datafram1=pd.DataFrame({
    'City':["Dubai","Abu Dhabi"],
    'Avg_Salary':[5000,4000]
})

Datafram2=pd.DataFrame({
     'City':["Alex","Cairo"],
    'Avg_Salary':[5000,4000]
})

new_df=pd.concat([Datafram1,Datafram2])

new_df.set_index("City")
```
